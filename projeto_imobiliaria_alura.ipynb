#Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

```python
dados["Valor"].mean()
```

```python
dados.groupby("Tipo").mean(numeric_only=True)
```

```python
dados.groupby("Tipo")["Valor"].mean()
```

```python
dados.groupby("Tipo")[["Valor"]].mean().sort_values("Valor").reset_index()
```

```python
df_preco_tipo = dados.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")

df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='darkred')
```

## Imóveis residenciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Box/Garagem', 'Loja Shopping/ Ct Comercial',
                      'Chácara', 'Loteamento/Condomínio',
                      'Sítio', 'Pousada/Chalé',
                      'Hotel', 'Indústria']
```

```python
dados.query('@imoveis_comerciais in Tipo')
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')

imoveis_residenciais.head()
```

```python
imoveis_residenciais['Tipo'].unique()
```

```python
imoveis_residenciais_preco_medio = imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

imoveis_residenciais_preco_medio.reset_index()
```

```python
imoveis_residenciais_preco_medio.plot(kind='barh', figsize=(14, 10), color='darkred')
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
imoveis_residenciais.Tipo.unique()
```

```python
imoveis_residenciais_percentual = imoveis_residenciais["Tipo"].value_counts(normalize=True).to_frame().reset_index().sort_values("Tipo")

imoveis_residenciais_percentual.columns = ["Tipo", "Percentual"]

imoveis_residenciais_percentual
```

```python
imoveis_residenciais_percentual_plotado = imoveis_residenciais_percentual.plot(kind='bar', figsize=(14, 10), color='darkgreen', xlabel="Tipos", ylabel="Percentual")

imoveis_residenciais_percentual_plotado
```

## Selecionando apenas os imóveis do tipo apartamento

```python
imoveis_residenciais.query('Tipo == "Apartamento"')
```

```python
imovel_apartamento = imoveis_residenciais.query('Tipo == "Apartamento"')
imovel_apartamento.head()
```

## Desafio

```python
inf_apartamento = dados.query('Tipo == "Apartamento"')['Quartos'].mean() # calcula a média de quartos no tipo de imóvel "Apartamento"

inf_apartamento
```

```python
len(dados['Bairro'].unique()) # mostra a quantidade de bairros presentes na base de dados
```

```python
bairros_maior_aluguel = dados.groupby('Bairro')['Valor'].mean().to_frame().sort_values('Valor', ascending=False).head()

bairros_maior_aluguel
# mostra os bairros que possuem a maior média de valor de aluguel na base de dados
```

```python
top_5_bairros_mais_caros = bairros_maior_aluguel.head()

top_5_bairros_mais_caros.plot(kind='barh', figsize=(14, 10), color='darkblue',

                              xlabel='Valor', ylabel='Bairro')
```
